```python
import statsmodels.api as sm
# fit the model
X = data[['LogVolume_BTC', 'SP_LogReturn']]
y = data['BTC_LogReturn_']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import seaborn as sns

# Residual plots
plt.figure(figsize=(8, 4))
sns.scatterplot(x=model.fittedvalues, y=model.resid)
plt.axhline(y=0, color='red', linestyle='--')
plt.title("Model Residual Plot")
plt.show()

# Q-Q plot for normality
plt.figure(figsize=(8, 4))
sm.qqplot(model.resid, line='s')
plt.title("Q-Q Plot")
plt.show()
```

```python
# Durbin-Watson test for autocorrelation
print("Durbin-Watson statistic: ", sm.stats.durbin_watson(model.resid))

# VIF for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Predicted values from the model
y_pred = model.predict(X)

# True values from the dataset
y_true = data['BTC_LogReturn_']

# Adjusted R-squared
r_squared = model.rsquared_adj

# AIC (Akaike Information Criterion)
aic = model.aic

# BIC (Bayesian Information Criterion)
bic = model.bic

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Print the metrics
print(f"Adjusted R²: {r_squared}")
print(f"AIC: {aic}")
print(f"BIC: {bic}")
print(f"RMSE: {rmse}")
print(f"MAPE: {mape}%")
```

# Cooks Distance

```python
import numpy as np
import statsmodels.api as sm

# Fit the model
X = sm.add_constant(data[['LogVolume_BTC', 'SP_LogReturn']])
model = sm.OLS(data['BTC_LogReturn_'], X).fit()

# Calculate residuals (u)
u = model.resid

# Calculate leverage values (h)
h = model.get_influence().hat_matrix_diag

# Calculate Cook's distance
p = X.shape[1]  # Number of predictors including intercept
n = X.shape[0]  # Number of observations
cooks_distance = (u**2 / (1 - h)) * (p / (n - p))

# Add Cook's distance to the data
data['cooks_distance'] = cooks_distance

# Filter out points with high Cook's distance
threshold = 4 / n  # Common threshold for Cook's distance
data = data[cooks_distance < threshold]
```

```python
data
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

model = sm.OLS(y, sm.add_constant(X)).fit()
influence = model.get_influence()
(c, p) = influence.cooks_distance

# Plot
plt.stem(np.arange(len(c)), c, markerfmt=",")
plt.axhline(4 / len(c), color='red', linestyle='--', label='Threshold (4/n)')
plt.title("Cook's Distance")
plt.xlabel('Observation Index')
plt.ylabel("Cook's Distance")
plt.legend()
plt.show()
```

```python
influential_point = np.argmax(c)  # or use a filter like c > 1
print(f"Most influential point index: {influential_point}")
data.iloc[influential_point]
```

```python
data_cleaned = data.drop(data.index[influential_point])
```

```python
# Durbin-Watson test for autocorrelation
print("Durbin-Watson statistic: ", sm.stats.durbin_watson(model1.resid))

# VIF for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
```

```python
# Predicted values from the model
y_pred = model1.predict(X)

# True values from the dataset
y_true = data_cleaned['BTC_LogReturn_']
# R-squared and Adjusted R-squared from the model
r_squared = model1.rsquared
adj_r_squared = model1.rsquared_adj

# Mean Squared Error (MSE)
mse = mean_squared_error(y_true, y_pred)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_true, y_pred)

# Mean Absolute Percentage Error (MAPE)
epsilon = 1e-8  # a small constant to avoid division by zero
mape = np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100

# Print the metrics
print(f"R-squared: {r_squared}")
print(f"Adjusted R-squared: {adj_r_squared}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")
```
